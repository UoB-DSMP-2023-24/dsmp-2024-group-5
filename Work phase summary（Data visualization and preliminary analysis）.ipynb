## 1.Import data and make preliminary observations

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = r'D:\桌面文件\Bristol\mini-project\vdjdb-2023-06-01\Vdjdb.xlsx'
df = pd.read_excel(file_path)
```

```python
print(df.head())
```

## 2. Data visualization
### I then performed some basic visualizations to explore the key features in the dataset

```python
#TCRV gene use frequency
plt.figure(figsize=(10, 6))
sns.countplot(data=df, y='v.segm', order=df['v.segm'].value_counts().index[:20])
plt.title('Top 20 TCR V Segment Usage')
plt.xlabel('Frequency')
plt.ylabel('V Segment')
plt.tight_layout()
plt.show()
```

The V gene, which has a high frequency of gene use, may indicate an immune response to a specific antigen or disease state. Therefore, the TCR can be subsequently grouped according to the use of the V gene.Same with J gene

```python
# Plot the length distribution of CDR3
df['cdr3_length'] = df['cdr3'].apply(len)

plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='cdr3_length', bins=range(min(df['cdr3_length']), max(df['cdr3_length']) + 1), kde=False)
plt.title('Distribution of CDR3 Lengths')
plt.xlabel('CDR3 Length')
plt.ylabel('Frequency')
plt.show()
```

TCRS of similar length may have similar structural properties that facilitate their recognition of similar epitopes. The CDR3 length is considered to be an important feature in the clustering process.

```python
# V and J gene pairing heat map
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='mhc.class')
plt.title('MHC Class Distribution')
plt.xlabel('MHC Class')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```
