## 4.  data visualization

```python
#gene
df = pd.DataFrame(data)

counts = df['gene'].value_counts()

# visualization
counts.plot(kind='bar')
plt.title('Number of TRA and TRB in gene')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()

print(counts)
```

```python
#complex.id
df = pd.DataFrame(data)

column_to_analyze = df['complex.id']

zero_count = (column_to_analyze == 0).sum()
non_zero_count = (column_to_analyze !=0).sum()

# visualization
plt.bar(['Zero', 'Non-Zero'],[zero_count, non_zero_count])
plt.title('Number of 0 and non-0 in complex.id')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()

print("Zero count:", zero_count)
print("Non-Zero count:", non_zero_count)
```

```python
#mhc.a 
df = pd.DataFrame(data)
counts = df['mhc.a'].value_counts()
plt.figure(figsize=(25,6))

# visualization
counts.plot(kind='bar')
plt.title('Counts of Each Category in mhc alpha chain')
plt.xlabel('Category')

plt.ylabel('Count')
plt.show()

print(counts)
```

```python
#mhc.b 
df = pd.DataFrame(data)
counts = df['mhc.b'].value_counts()
plt.figure(figsize=(25,6))

# visualization
counts.plot(kind='bar')
plt.title('Counts of Each Category in mhc beta chain')
plt.xlabel('Category')

plt.ylabel('Count')
plt.show()

print(counts)
```

```python
#mhc.class frequency
df = pd.DataFrame(data)
counts = df['mhc.class'].value_counts()
plt.figure(figsize=(25,6))

# visualization
counts.plot(kind='bar')
plt.title('numbers of MHC-I and MHC-II in mhc.class')
plt.xlabel('Category')

plt.ylabel('Count')
plt.show()

print(counts)
```
